```python
import numpy as np
import matplotlib.pyplot as plt

duration = 4  # длительность сигнала в секундах
amplitude = 0.4  # амплитуда  (в пределах: +-1.0)

fs = 44000 # частота дискретизации

timeSamples = np.arange(np.ceil(duration * fs)) / fs

signal1 = amplitude * np.sin(2 * np.pi * 293 * timeSamples) #ре 1 окта
signal2 = amplitude * np.sin(2 * np.pi * 329 * timeSamples) #ми 1 окта
signal3 = amplitude * np.sin(2 * np.pi * 277 * timeSamples) #до-диез 1 окта
signal4 = amplitude * np.sin(2 * np.pi * 246 * timeSamples) #си малой окта
signal5 = amplitude * np.sin(2 * np.pi * 233 * timeSamples) #ля-диез малой окта
signal6 = amplitude * np.sin(2 * np.pi * 185 * timeSamples) #фа-диез малой окта
signal7 = amplitude * np.sin(2 * np.pi * 246 * timeSamples) #си малой окта длинная

signalSumm = signal1 + signal2 + signal3 + signal4 + signal5 + signal6 + signal7


plt.plot(signalSumm[:500])
plt.title("sine tone")
plt.xlabel("time / milliseconds")
```

```python
N = duration * fs

spector = np.fft.rfft(signalSumm)
x_axis = np.fft.rfftfreq(N, 1/fs)

plt.plot(x_axis, np.abs(spector))
plt.show()
```

```python
new_sig = np.fft.irfft(spector)
plt.plot(new_sig[:500])
plt.show()
```
